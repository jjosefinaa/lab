```python
import pyvisa
from pyvisa import VisaIOError
```

```python
rm = pyvisa.ResourceManager()

for s in rm.list_resources():
    try:
        device = rm.open_resource(s)
        print(s, device.query('*IDN?').strip())
    except VisaIOError:
        continue
```

```python
rm = pyvisa.ResourceManager()
device = rm.open_resource('USB0::0x0699::0x034C::C010119::INSTR')  # where the FG is

print("Connected to:", device.query("*IDN?").strip())
```

```python
# to change amplitude

amp = 1.5 #Vpp

device.write(f'SOUR1:VOLT {amp}')

amp_actual = device.query('SOUR1:VOLT?')

print("Amplitude:", amp_actual.strip(), "Vpp")
```

```python
# to change frequency

freq = 100 #Hz

device.write(f'SOUR1:FREQ {freq}')

freq_actual = device.query("SOUR1:FREQ?")

print("Frequency:", freq_actual.strip(), "Hz")
```

```python
# to change waveform
wf = {0: 'SINusoid', 1: 'SQUare', 2: 'RAMP'}

device.write(f'SOUR1:FUNCtion {wf[2]}')

wf_actual = device.query("SOUR1:FUNCtion?")

print("Waveform:", wf_actual.strip())
```

```python
# output of device (?)
device.write('OUTPUT1:STATE ON')  # ON or OFF

out = device.query('OUTP1?')  # channel 1 ON or OFF
print("Output:", out.strip())
```
